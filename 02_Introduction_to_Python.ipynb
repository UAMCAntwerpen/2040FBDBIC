## Operators

```python
fruits = ["mango", "orange", "lemon"]

# Membership operators
print("strawberry" in fruits)
print("strawberry" not in fruits)
print("")

# Logical operators (or, and, not)
print("strawberry" in fruits and "cocos" in fruits)
print("strawberry" in fruits or "cocos" in fruits)
print("strawberry" in fruits or "cocos" not in fruits)
print("strawberry" not in fruits or "cocos" in fruits)
print("strawberry" not in fruits and "cocos" in fruits)
print("")

# Comparison operators
print(3 > 4)
print(3 < 4)
print( 3 <= 4)
print(3 == 3)
print(3 != 4)
print("")
```

## For loops

```python
fruits = ["apple", "mango", "orange"]
for fruit in fruits:
  print(fruit)

for i in range(0, 10, 1):
  print(i, i + 10)

for i in range(0, len(fruits)):
  print(fruits[i])

for i in range(0, len(fruits)):
  fruits[i] = fruits[i] + "s"
print(fruits)
```

## While loops

```python
target = 10
index = 1
while index <= target:
  print(index, target)
  index += 1
```

## Interrupting loops

```python
for i in range(10):
  if i == 5:
    continue
  print(i)

print("")
i = 0
while i < 100:
  i += 1
  if i == 5: break
  if i == 3: continue
  print(i)
```
